```python
# 함수의 데코레이터 사용
def test(func):
    def wrapper():
        print("start !!!")
        func()
        print("end !!!")
    return wrapper

# @데코레이션을 붙여 함수 생성
@test
def hello():
    print("hello !!")
    
# 함수 호출
hello() #-> == test(hello())
```

```python
#객체를 생성하는 함수를 작성
def create_student(name, korean, math,english, science):
    return {"name": name , "korean": korean, 
            "math": math, "english": english, "science": science}

#학생을 처리하는 함수 선언
def student_get_sum(student): #점수 합
    return student['korean'] + student['math'] \
        +student['english'] + student['science']

def student_get_avg(student): #평균
    return student_get_sum(student) / 4

def student_to_string(student): #개인 학생 정보 출력
        return "{}\t{}\t{}".format(
            student["name"], student_get_sum(student), 
            student_get_avg(student))
```

```python
class Student:   #클래스 정의
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean  #클래스 변수 명칭: 멤버
        self.math = math
        self.english = english
        self.science = science
    
    def get_sum(self): #클래스 함수 명칭: 메서드
        return self.korean + self.math + self.english + self.science
    
    def get_avg(self):
        return self.get_sum() / 4
    
    def to_string(self):
        return "{}\t{}\t{}".format(
            self.name, self.get_sum(), self.get_avg())
```

```python

class Student:   #클래스 정의
    def __init__(self, name, korean, english):
        self.name = name
        self.korean = korean  #클래스 변수 명칭: 멤버
        self.english = english
    
    def get_sum(self): #클래스 함수 명칭: 메서드
        return self.korean + self.english
    
    def get_avg(self):
        return self.get_sum() / 4
    
    def __str__(self):
        return "{}\t{}\t{}".format(
            self.name, self.get_sum(), self.get_avg())

    def __eq__(self, value):
        return self.get_sum() == value.get_sum()
    def __gt__(self, value):
        return self.get_sum() > value.get_sum()
```

```python
# 프라이빗 변수(__변수이름)와 게터 세터
# 프라이빗 변수: 외부에서 클래스 인스턴스의 값을 수정하지 못하도록 함

class Test:
    def __init__(self,score):
        self.score = score
        
    def get_score():
        return self.score
    
    def set_score(self, value):
        self.score = value
        
a= Test(90)
print(a.score)
a.score = 100
print(a.score)
```
