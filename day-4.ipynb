```python
def sum_f(start=0 ,stop=100, step=1):
    total = 0
    for i in range(start, stop+1, step):
        total += i
    return total

print("0~100 합:", sum_f())
print("0~100 까지 짝수의 합:", sum_f(step=2))
print("0~100 까지 홀수의 합:", sum_f(start=1, step=2))
print("0~50 까지 홀수의 합:", sum_f(start=1, step=2))
```

```python
#계산기
def calc_f(a,b,n):
    lst=[]

    if n == '+':
        r = str(a + b)
    elif n == '-':
        r = str(a - b)
    elif n == '*':
        r = str(a * b)
    elif n == '/':
        r = str(a / b)
    else:
        r=0

    if r==0:
        print("부호 오류")
    else :
        print("{} {} {} = {}".format(a,n,b,r))

calc_f(10,20,"+")
```

```python
#재귀함수

def fac_f(n): # factorial
    if n == 0:
        return 1
    else:
        return n*fac_f(n-1)
        
fac_f(10)
```

```python
def power(item):
    return item*item
def under_3(item):
    return item < 3

# map (함수명, 리스트) -> 리스트의 각각의 값을 함수에 적용
# filter(함수명, 리스트) -> 함수에 조건식으로 return
```

```python
# lambda 매개변수 : 리턴 값
power = lambda x: x*x
under_3 = lambda x: x<3

input_list = [1,2,3,4,5]

output = map(power, input_list)
print(list(output))

print(list(filter(under_3, input_list)))
```

```python
input_list = [1,2,3,4,5]

output = map(lambda x: x*x, input_list)
print(list(output))

print(list(filter(lambda x: x<3, input_list)))
```

```python
# 제너레이터 함수 작성: 함수 내부에 yield 키워드 사용
# 함수를 부를때마다 yield 까지만 순서대로 실행 : next(f)사용
def gen_test():
    print('First ')
    yield 1
    
    print('Second ')
    yield 2
    
    print('Third ')

o_func = gen_test()   

print("main start ")
a = next(o_func)
print(a)

print("main 111 start ")
a = next(o_func)
print(a)

print("main 222 start ")
a = next(o_func)
print(a)
```

```python
# 실행오류: try ~ except ~ else ~ finally
# 실행 오류가 발생할 수 있는 코드 작성

try: #코드
    input_num = int(input("정수 입력: "))
except: #에러 시 실행
    print("정수가 입력되지 않음")
else: #정상작동시 실행
    print("원의 반지름: ", input_num)
    print("원의 둘레: ", input_num * 2 * 3.14)
finally: #무조건 실행
    print("program end !!")
```
